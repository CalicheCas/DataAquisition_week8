```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Get  Car data

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data", sep="\s+")
df.head()
```

### Add column names to data frame

```python
columns = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model year", "origin", "car name"]
```

```python
df.columns = columns
df.head()
```

### Replace '?' with 0.00

```python
df.loc[df['horsepower'] == "?", 'horsepower'] = 0.0
```

### Replace Origin column 1 to USA

```python
df.loc[df['origin'] == 1, 'origin'] = "USA"
```

### Replace Origin column 2 to Europe

```python
df.loc[df['origin'] == 2, 'origin'] = "Europe"
```

### Replace Origin column 3 to Asia

```python
df.loc[df['origin'] == 3, 'origin'] = "Asia"
```

### What are the most econopic cars from USA, Asia and Eurpe? how does acceleration impacts fuel economy

```python
usa = df[df.origin == 'USA'].nlargest(20, columns='mpg')
usa
```

```python
sns.lmplot("mpg", "acceleration", data=usa, fit_reg=False)
```

```python
asi = df[df.origin == 'Asia'].nlargest(20, columns='mpg')
asi
```

```python
sns.lmplot("mpg", "acceleration", data=asi, fit_reg=False)
```

```python
eu = df[df.origin == 'Europe'].nlargest(20, columns='mpg')
eu
```

```python
sns.lmplot("mpg", "acceleration", data=eur, fit_reg=False)
```
